```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('numerical.csv')
data.head(5)
```

```python
sns.displot(data['customer_lifetime_value'], bins=50)
sns.displot(data['income'], bins=50)
sns.displot(data['monthly_premium_auto'], bins=50)
sns.displot(data['months_since_last_claim'], bins=50)
sns.displot(data['months_since_policy_inception'], bins=50)
sns.displot(data['number_of_open_complaints'], bins=50)
sns.displot(data['number_of_policies'], bins=50)
sns.displot(data['total_claim_amount'], bins=50)
plt.show()
```

#### none look like normal distribution, a few (total claim amount, number of open complaints, monthly premium auto and customer lifetime value) like (more or less) exponential distribution

```python
data_log = np.log(data+1)
data_log
# data+1 to avoid -inf where value is 0
```

```python
sns.displot(data_log['customer_lifetime_value'], bins=50)
sns.displot(data_log['income'], bins=50)
sns.displot(data_log['monthly_premium_auto'], bins=50)
sns.displot(data_log['months_since_last_claim'], bins=50)
sns.displot(data_log['months_since_policy_inception'], bins=50)
sns.displot(data_log['number_of_open_complaints'], bins=50)
sns.displot(data_log['number_of_policies'], bins=50)
sns.displot(data_log['total_claim_amount'], bins=50)
plt.show()
```

#### customer lifetime value looks a bit more like normal distribution, so does total claim amount

```python
data_sqrt = np.sqrt(data)
data_sqrt
```

```python
sns.displot(data_sqrt['customer_lifetime_value'], bins=50)
sns.displot(data_sqrt['income'], bins=50)
sns.displot(data_sqrt['monthly_premium_auto'], bins=50)
sns.displot(data_sqrt['months_since_last_claim'], bins=50)
sns.displot(data_sqrt['months_since_policy_inception'], bins=50)
sns.displot(data_sqrt['number_of_open_complaints'], bins=50)
sns.displot(data_sqrt['number_of_policies'], bins=50)
sns.displot(data_sqrt['total_claim_amount'], bins=50)
plt.show()
```

#### no big difference, only total claim amount starts looking more like normal ditribution
