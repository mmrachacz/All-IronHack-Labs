```python
data1 = data.pivot_table(index = ['CustomerID'], values = ['amount_spent'], aggfunc = {'amount_spent':sum})
data1.reset_index()
```

```python
# Calculate the 0.95 and 0.75 quantiles of the 'total_spent' column
quantiles = data1['amount_spent'].quantile([0.95, 0.75])
quantiles
```

```python
binnames = ['regular', 'preferred', 'VIP']
data1['CustomerType'] = pd.cut(data1['amount_spent'], bins=[0,quantiles[0.75],quantiles[0.95],100000], labels = binnames)
data1.head(10)
```
