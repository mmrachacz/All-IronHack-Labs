## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?
customer_data = orders.groupby(['CustomerID','Country']).agg({'amount_spent':sum})
customer_data
```

```python
# Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

# Calculate the 0.95 and 0.75 quantiles of the 'total_spent' column
quantiles = customer_data['amount_spent'].quantile([0.95, 0.75])
quantiles
```

```python
labels=['Long Tail','Preferred','VIP']
customer_data['customer_type'] = pd.cut(customer_data['amount_spent'], bins=[-np.inf, quantiles[0.75], quantiles[0.95], np.inf], labels=labels)
customer_data
```

```python
# Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?
vip_customer = customer_data[customer_data['customer_type']=='VIP']
preferred_customer = customer_data[customer_data['customer_type']=='Preferred']
preferred_customer
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
vip_country = vip_customer.groupby(['Country']).agg({'amount_spent':sum}).sort_values('amount_spent', ascending=False)
vip_country
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
#creating preferred list

preferred_country = preferred_customer.groupby(['Country']).agg({'amount_spent':sum}).sort_values('amount_spent', ascending=False)

#Combining both


vip_and_preferred = pd.merge(left = vip_country,
                                 right = preferred_country,
                                 how = 'inner', 
                                 left_on = "Country", 
                                 right_on= "Country")

vip_and_preferred
```
