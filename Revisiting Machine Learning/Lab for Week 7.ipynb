### Downsampling:

```python
category_0_undersampled = resample(category_0, 
                                   replace=False, 
                                   n_samples = len(category_1))
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

### Upsampling:

```python
category_1_oversampled = resample(category_1, 
                                  replace=True, 
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
y = data_downsampled['TARGET_B']
X = data_downsampled.drop(['TARGET_B'], axis = 1)

num = X.select_dtypes(np.number)
cat = X.select_dtypes(object)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)

X_train_cat[['STATE', 'CLUSTER', 'HOMEOWNR', 'GENDER', 'WEALTH1', 'DATASRCE', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A', 'DOMAIN_B']] = X_train_cat[['STATE', 'CLUSTER', 'HOMEOWNR', 'GENDER', 'WEALTH1', 'DATASRCE', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A', 'DOMAIN_B']].astype(str)
X_test_cat[['STATE', 'CLUSTER', 'HOMEOWNR', 'GENDER', 'WEALTH1', 'DATASRCE', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A', 'DOMAIN_B']] = X_test_cat[['STATE', 'CLUSTER', 'HOMEOWNR', 'GENDER', 'WEALTH1', 'DATASRCE', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A', 'DOMAIN_B']].astype(str)

scaler = MinMaxScaler()
X_train_num_scaled = pd.DataFrame(scaler.fit_transform(X_train_num), columns=X_train_num.columns)
X_test_num_scaled = pd.DataFrame(scaler.transform(X_test_num), columns=X_train_num.columns)

encoder = OneHotEncoder(drop='first').fit(X_train_cat) 
cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)
X_train_cat_encode = pd.DataFrame(encoder.transform(X_train_cat).toarray(),columns=cols)
X_test_cat_encode = pd.DataFrame(encoder.transform(X_test_cat).toarray(),columns=cols)

X_train  = pd.concat([X_train_num_scaled, X_train_cat_encode], axis=1)
X_test  = pd.concat([X_test_num_scaled, X_test_cat_encode], axis=1)
```

```python
classification = LogisticRegression(max_iter = 500)
classification.fit(X_train, y_train)
classification.score(X_train, y_train)

predictions = classification.predict(X_test)
predictions
classification.score(X_test, y_test)
```
