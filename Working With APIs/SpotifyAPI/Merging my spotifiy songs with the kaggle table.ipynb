```python
import pandas as pd
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
my_data = pd.read_csv('songs_with_feats_and_clusters.csv')
kaggle_data = pd.read_csv('tracks.csv')
```

```python
scaler = pickle.load(open('minmax.pkl','rb'))
```

```python
display(my_data.head(1))
display(kaggle_data.head(1))
display(my_data.shape)
display(kaggle_data.shape)
```

```python
print(my_data.columns)
print(kaggle_data.columns)
```

```python
kaggle_data = kaggle_data[['name', 'artists', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms', 'time_signature', 'id',
       ]]
kaggle_data = kaggle_data.rename({'name': 'song', 'artists': 'artist'}, axis=1)
kaggle_data.head(1)
```

```python
def clean_alt_list(list_):
    list_ = list_.replace("'", "")
    list_ = list_.replace('[', '')
    list_ = list_.replace(']', '')
    return list_

kaggle_data['artist'] = kaggle_data['artist'].apply(clean_alt_list)
kaggle_data.head(1)
```

```python
display(my_data.head(1))
display(kaggle_data.head(1))
```

```python
songs_num = kaggle_data.select_dtypes(include='number')
songs_num
```

```python
songs_num_norm = pd.DataFrame(scaler.transform(songs_num), columns=songs_num.columns)
songs_num_norm
```

```python
kmeans = KMeans(n_clusters=13)
kmeans = kmeans.fit(songs_num_norm)

clusters = kmeans.predict(songs_num_norm)
len(clusters)
```

```python
pickle.dump(kmeans, open('kmeans.pkl','wb'))
```

```python
pd.Series(clusters).value_counts().sort_index()
```

```python
kaggle_data.loc[:,"cluster"] = kmeans.labels_
kaggle_data
```

```python
audio_feats_final = pd.concat([my_data, kaggle_data], axis=0)
audio_feats_final
```

```python
audio_feats_final = audio_feats_final.drop(['uri'], axis = 1)
audio_feats_final
```

```python
audio_feats_final.to_csv('kaggle_and_my_songs.csv', index = False)
```
