# Carbon Footprint of Websites (eg. IronHack)

```python
flattened_data = json_normalize(results)
flattened_data
```

# Bible verses

```python
response = requests.get('https://bible-api.com/romans 12:1-2,5-7,9,13:1-9&10')
results = response.json()
# results
```

```python
flattened_data = json_normalize(results)
flattened_data.T
```

```python
verses = flattened_data.verses[0]
verses_norm = pd.json_normalize(verses)
verses_norm
```
