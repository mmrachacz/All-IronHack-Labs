```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
# we need Track-Name, Track-Artist, Track-URI
```

# Why not start with Jans example:

```python
secrets_file = open("secrets.txt","r")

string = secrets_file.read()

secrets_dict={}

for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
        
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['cid'],
                                                           client_secret=secrets_dict['csecret']))
```

```python
playlist = sp.user_playlist_tracks("spotify", "4rnleEAOdmFAbRcNCgZMpY")
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3))
    return tracks
```

```python
%%time
all_tracks = get_playlist_tracks("4rnleEAOdmFAbRcNCgZMpY")
len(all_tracks)
```

```python
songs = pd.DataFrame({"song":song,
                        "artist":artists,
                          "uri":uri,
                      })
songs
```
