# Checking for Multicolliniarity:

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlations_matrix = fifa_num.corr()
display(correlations_matrix)
# fig, ax = plt.subplots(figsize=(50,50)) # command to make the sizes of the fields bigger 
# sns.set(font_scale=1.6)
# sns.heatmap(correlations_matrix, annot=True)
# plt.show()
```

```python
# compute the correlation matrix
corr_matrix = fifa_num.corr()

# create a mask to hide the lower triangle of the heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# filter the correlation matrix to show only high correlations
high_corr = corr_matrix[abs(corr_matrix) > 0.9]

fig, ax = plt.subplots(figsize=(50,50))

# create the heatmap with high correlation values only
sns.heatmap(high_corr, cmap='coolwarm', mask=mask, annot=True)

# show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = fifa_num.corr()
high_corr_cols = set()
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > 0.9:
            colname = corr.columns[i]
            high_corr_cols.add(colname)

fifa_num = fifa_num.drop(high_corr_cols, axis=1)
fig, ax = plt.subplots(figsize=(50,50))
sns.set(font_scale=1.4)
# Create a heatmap
sns.heatmap(fifa_num.corr(), annot=True)
print(high_corr_cols)
plt.show()


```

# Scaling the Data:

```python
# X,y-split
y = fifa_num['ova'] # define the label
X = fifa_num.drop(['ova'], axis = 1) #define features
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler 
```

```python
# Normalizing data: make data range from 0 - 1, instead of from min to max
min_max_transformer = MinMaxScaler().fit(X) # train the scaler with the values of X
fifa_num_normalized = min_max_transformer.transform(X) # storing the transformed data in a variable and transforming it
#print(x_normalized.shape)
fifa_num_normalized = pd.DataFrame(fifa_num_normalized, columns=X.columns)
fifa_num_normalized
```

```python
fifa_num_normalized.describe().T
```

```python
fifa_cat.columns
```

```python
fifa_cat2 = fifa_cat.copy()
```

```python
fifa_cat2 = fifa_cat2.drop(['position'], axis = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from seaborn import load_dataset
from sklearn.compose import make_column_transformer

encoder = OneHotEncoder(drop='first').fit(fifa_cat2) 

cols = encoder.get_feature_names_out(input_features=fifa_cat2.columns)

fifa_cat_encode = pd.DataFrame(encoder.transform(fifa_cat2).toarray(),columns=cols)

display(fifa_cat_encode.head())
fifa_cat_encode.shape
```

# Training and Predicting:

```python
X = pd.concat([fifa_num_normalized, fifa_cat_encode], axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
import matplotlib
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# X_summary = pd.DataFrame(X.isna().sum(), columns=['Number of NaN'])
# def print_full(df):
#     pd.set_option('display.max_rows', len(df))
#     print(df)
#     pd.reset_option('display.max_rows')

# X_summary.apply(print_full)
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```

```python
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
rmse
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mae = mean_absolute_error(y_test, predictions_test)
print(mae)
```
