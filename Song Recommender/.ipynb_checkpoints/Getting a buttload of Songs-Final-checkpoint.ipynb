# Why not start with Jans example:

```python
secrets_file = open("secrets.txt","r")

string = secrets_file.read()

secrets_dict={}

for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
        
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['cid'],
                                                           client_secret=secrets_dict['csecret']))
```

```python
playlist = sp.user_playlist_tracks("spotify", "4rnleEAOdmFAbRcNCgZMpY")
```

```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3))
    return tracks
```

# Clustering:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pickle
```

```python
# splitting cat and num columns
songs_num = song_with_feats.select_dtypes(include='number')
songs_cat = song_with_feats.select_dtypes(include='object')
```

```python
transformer = MinMaxScaler().fit(songs_num)
songs_num_norm = pd.DataFrame(MinMaxScaler().fit_transform(songs_num), columns=songs_num.columns)
songs_num_norm
pickle.dump(transformer, open('minmax.pkl', 'wb'))
```

```python
kmeans = KMeans(n_clusters=13)
kmeans = kmeans.fit(songs_num_norm)
```

```python
import pickle

# pickle.dump(kmeans, open('kmeans.pkl', 'wb'))
# songs_num_norm.to_pickle('songs_num_norm.p')
```

# assigning the clusters:
clusters = kmeans.predict(songs_num_norm)
# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()

```python
kmeans.inertia_ 
```

```python
# looking for optimal k
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    )
    kmeans.fit(songs_num_norm)
    inertia.append(kmeans.inertia_)
```

```python
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
#plt.xticks(np.arange(min(K), max(K)+1, 1.0))
```

```python
# no "elbow" point distinguishable... checking for silhouette_score
```

```python
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(songs_num_norm)
    silhouette.append(silhouette_score(songs_num_norm, kmeans.predict(songs_num_norm)))
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
```

```python
# maybe k=13, because its not the absolute maximum in silhuette score but a partial in the imediate area
kmeans = KMeans(n_clusters=13)
kmeans.fit(songs_num_norm)
# assigning the clusters:
clusters = kmeans.predict(songs_num_norm)
# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

##### I am going with 13 Clusters because:
##### - in the silhuette score it i a highpoint in its area
##### - also, i have more than 17000 songs, which are all quite different, so more clusters will be more true
##### - and the songs are pretty evenly distributed within the clusters...

# Building the recommender itself in another Notebook

```python
top100 = pd.read_csv('songs.csv')
```

```python
input_song = [input('Enter your favorite songs\n> ').strip()]
```

```python
import random

song_lst = top100['song'].tolist()
favorites = top100[top100.song.isin(input_song)]
fav_lst = favorites.song.tolist()
track_feats = pd.DataFrame()
# print(fav_lst)

for i in favorites:
    if not not fav_lst:
        print(random.sample(song_lst, k=1))
    if not fav_lst:
#         track_feats = pd.DataFrame()
        for a in input_song:
            track = sp.search(q=a)
            track_uri = [track['tracks']['items'][0]['uri']]
        #     print(v)
        #     print(track_uri)
        for b in track_uri:
            track_feats = track_feats.append((sp.audio_features(b)),ignore_index=True)
            track_feats = track_feats.select_dtypes(include='number')
            track_feats = pd.DataFrame(MinMaxScaler().fit_transform(track_feats), columns=track_feats.columns)
            track_cluster = kmeans.predict(track_feats).tolist()
        #     display(track_feats)
        #     display(track_cluster)
        for c in track_cluster:
            rec = song_with_feats[(song_with_feats['cluster'] == c)].sample(1)
            display(rec[['song','artist']])
```
