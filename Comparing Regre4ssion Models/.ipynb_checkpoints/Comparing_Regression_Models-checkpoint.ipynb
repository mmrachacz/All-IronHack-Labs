```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
data_num = pd.read_csv('numerical.csv')
data_cat = pd.read_csv('categorical.csv')
```

```python
data = pd.concat((data_cat,data_num ), axis = 1)
```

```python
# X,y Split (Total Claim Amount as Target)
X = data.drop(['total_claim_amount'], axis=1)
y = data['total_claim_amount']
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Separating X_train and X_test into numerical and categorical
X_train_num = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
```

```python
# ordinal encoding the 2 categorical columns that will not be OneHotEncoded
# checking for unique values to determine the 'classes'
# display(X.coverage.unique())
# display(X.vehicle_size.unique())
X_train_cat["coverage"] = X_train_cat["coverage"].map({"Basic" : 0, "Extended" : 0.5, "Premium" : 1}).astype(str)
X_train_cat["vehicle_size"] = X_train_cat["vehicle_size"].map({"Small" : 0, "Medsize" : 0.5, "Large" : 1}).astype(str)
X_test_cat["coverage"] = X_test_cat["coverage"].map({"Basic" : 0, "Extended" : 0.5, "Premium" : 1}).astype(str)
X_test_cat["vehicle_size"] = X_test_cat["vehicle_size"].map({"Small" : 0, "Medsize" : 0.5, "Large" : 1}).astype(str)
```

```python
# k neighbor prediction on testing data really bad, so checking for a better numbr of k
scores = []
for k in range(2,25):
    model = KNeighborsRegressor(n_neighbors=k, weights='uniform', algorithm='auto', p=1)
    model.fit(X_train_treated, y_train)
    scores.append(model.score(X_test_treated, y_test))
```

```python
x = X_train_treated
y = X_test_treated

def multiple_models(x,y):
    # linear regression
    lm = linear_model.LinearRegression()
    lm.fit(x,y_train)
    predictions = lm.predict(x)
    LM_train_score = r2_score(y_train, predictions)
    predictions_test = lm.predict(y)
    LM_test_score = r2_score(y_test, predictions_test)
    
    # k neighbor
    KNN = KNeighborsRegressor(n_neighbors=6, weights='uniform', algorithm='auto', p=1, n_jobs=1)
    KNN.fit(x, y_train)
    KNN_train_score = KNN.score(x, y_train)
    KNN_test_score = KNN.score(y, y_test)
    
    return LM_test_score, KNN_test_score
```

#### Prediction results for LM are higher values than for KNN (where the optimum lies somewhere 0.54)

```python
def multiple_predictions(models, data):
    predictions = []
    for model in models:
        prediction = model.predict(data)
        predictions.append(prediction)
    return predictions

```

```python
# Example usage:
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Create some example models
model1 = LinearRegression().fit(data)
model2 = RandomForestRegressor(n_estimators=100).fit(data)

# Combine the models into a list
models = [model1, model2]

# Input data for prediction
data = X_train_treated

# Call the function to get predictions from both models
predictions = multiple_predictions(models, data)

# Print the predictions
print(predictions)
```
